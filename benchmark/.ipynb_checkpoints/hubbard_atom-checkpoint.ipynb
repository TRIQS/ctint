# Benchmark Hubbard Atom

```python
# Parameters
U = 2.0
beta = 1.0
mu = U/2.0
n_iw = 20
n_tau = 1001

# The Hamiltonian of the Hubbard Atom
h_int = U * n('up',0)*n('down',0)

# The alpha shift : alpha[block][index,s]
delta = 0.3
diag = 0.5 + delta
odiag = 0.5 - delta
alpha = [ [[diag,odiag]], [[odiag,diag]] ]

# The Green function structure
gf_struct = { 'up':[0], 'down':[0] }

# Construct the CTINT
from ctint import SolverCore
S = SolverCore(beta = beta,
               gf_struct = gf_struct,
               n_iw = n_iw,
               n_tau = n_tau )

# Initialize the non-interacting Green's function
for name ,g0 in S.G0_iw:
    g0 << inverse(iOmega_n + mu);
    
# Run the QMC
S.solve(h_int=h_int,
        alpha = alpha,
        n_cycles = 10000000,
        length_cycle = 50,
        n_warmup_cycles = 10000,
        measure_M_tau = True,
        n_iw_M3=n_iw,
        n_tau_M3=n_tau,
        measure_M3pp_tau = True,
        measure_M3ph_tau = True,
        post_process = True )

# Print average sign
print S.average_sign
```

# $\Sigma(i\omega)$

```python
# Plot the difference to the exact result
Sigma_at = GfImFreq( indices=[0], beta = beta, name = 'exact', n_points = n_iw)
Sigma_at << U/2.0 + U * U / 4.0 * inverse(iOmega_n)

oplot(S.Sigma_iw['up'] - Sigma_at, '-o')
```

# $G(i\omega)$

```python
# Plot the difference to the exact result
G_at = GfImFreq(indices=[0], beta=beta, n_points=20)
G_at << inverse(iOmega_n - U**2/4. * inverse(iOmega_n))

oplot(G_at - S.G_iw['up'])
```
